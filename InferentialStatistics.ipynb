A school wants to compare the effectiveness of 2 diffeent teaching methods.
Method A was used to each one class of students aand their marks after the exam were

CLASS A

75, 80, 82, 78, 85

Method B was used to teach another  clss and their marks were

70, 72, 75, 68, 74

At a 5% significance level, test whether there is a statistically significant difference between the average marks of students taught using method A & B

Task:

1. Define the null Hypothesis and Alternate Hypothesis

2. Identify the appropriae statistical test for the problem and justify the answer

3. perform the hypothesis test using python code

4. Calculate the t-statistics and p value

Note: t-statistics is a value calculated in t test that tells us how large the difference between 2 groups is compared to the variation in the data
if t-statistics value is close to 0, means that the groups are very similar
if it is very small like 0.5, 0.4,... means dfference may be due to chance
if it is large positive value, that means group 1 mean is much higher.
if it is large negative value, that means group 2 mean is much higher

5. Compare the p value with significance level

6. State your decision and intepret the result in the context of teaching methods.

```python
from scipy.stats import ttest_ind
class_A=[75,80,82,78,85]
class_B=[70,72,75,68,74]
t_statistic,p_value=ttest_ind(class_A, class_B)
print("T-statistic:", round(t_statistic,3))
print("P-Value    :",round(p_value,4))
alpha=0.05
if p_value < alpha:
  print("\nDecision : Reject the Null Hypothesis (Ho)")
  print("Conclusion: There is a statistically significant difference between the average marks of two classes")
else:
  print("\nDecision : Accept the Null Hypothesis (Ho)")
  print("Conclusion: There is no statistically significant difference between the average marks of two classes")
```
