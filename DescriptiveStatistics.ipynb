45, 50, 55, 60, 60, 65, 70, 75, 80, 90

Find the mean, median, mode, range, variance and Standard Deviation.

```python
import numpy as np
from statistics import mean, median, mode
data=[45,50,55,60,60,65,70,75,80,90]
print("Mean: ",mean(data))
print("Median: ",median(data))
print("Mode: ",mode(data))
print("Range: ",max(data) - min(data))
print("Population Variance:",np.var(data))
print("Population Standard Deviation:",np.std(data))
```
