```python
n=int(input("Student record\nChoose an option from below\n1. Add Student\n2. Display All Students\n3. Search Student by Roll Number\n4. Find Topper\n5. Display Students who Passed\n6. Count Students Failed\n7. Display Subject-wise Highest Marks\n8. Exit\n")
)
while n!=8:
  if n==1:
    print("Add Student")
  elif n==2:
    print("All Students")
  elif n==3:
    print("Search Student by Roll Number")
  elif n==4:
    print("Find Topper")
  elif n==5:
    print("Display Students who Passed")
  elif n==6:
    print("Count Students Failed")
  elif n==7:
    print("Display Subject-wise Highest Marks")
  elif n==8:
    break
  else:
    print("Invalid Option")
n=int(input("Enter the choice"))
```
